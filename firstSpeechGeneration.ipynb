<h1>Audio course</h1>

```python
from transformers import pipeline
pipe=pipeline("text-to-speech",model="suno/bark-small")
text="Hello this first text to speech model of mine, I ma so happy if it  works, wow it worked"
```

```python
output=pipe(text)
output
```

```python
from IPython.display import Audio
Audio(output["audio"],rate=output['sampling_rate'],autoplay=True)
```

<h1>Music Generating</h1>

```python
music_pipe = pipeline("text-to-audio", model="facebook/musicgen-small")
text = "Under neon lights, we dance so free,A rhythm in the night, just you and me."
forward_params = {"max_new_tokens": 512}

output = music_pipe(text, forward_params=forward_params)
Audio(output["audio"][0], rate=output["sampling_rate"])
```
